### Desarrollo Analítico: Método de la Transformada Inversa

Sea $X$ la variable aleatoria que representa el peso de una tina en kilogramos. De acuerdo con los datos de la empresa TIBASA, $X$ sigue una distribución triangular definida por los siguientes parámetros:

* **Límite inferior ($a$):** 190
* **Moda ($c$):** 210
* **Límite superior ($b$):** 230

Para aplicar el método de la transformada inversa, primero determinamos el valor de la Función de Distribución Acumulada (FDA) en la moda, lo cual nos indicará el punto de quiebre para nuestra probabilidad acumulada:
\begin{equation}
    F(c) = \frac{c - a}{b - a} = \frac{210 - 190}{230 - 190} = \frac{20}{40} = 0.5
\end{equation}

El método consiste en igualar la FDA, $F(x)$, a un número aleatorio uniforme $U \sim \text{U}(0,1)$ y despejar la variable $x$. Dado que la distribución triangular es una función definida a trozos, dividiremos el análisis en dos casos.

###Caso 1: Rama izquierda ($U \leq 0.5$)

Para el intervalo $a \leq x \leq c$, la función de distribución acumulada está dada por:
\begin{equation}
    F(x) = \frac{(x - a)^2}{(b - a)(c - a)}
\end{equation}

Igualamos $F(x)$ a $U$ y sustituimos los parámetros de nuestro problema:
\begin{equation}
    U = \frac{(x - 190)^2}{(230 - 190)(210 - 190)} = \frac{(x - 190)^2}{(40)(20)} = \frac{(x - 190)^2}{800}
\end{equation}

Procedemos a despejar $x$:
\begin{align}
    (x - 190)^2 &= 800 \cdot U \nonumber \\
    x - 190 &= \sqrt{800 \cdot U} \nonumber \\
    x &= 190 + \sqrt{800 \cdot U}
\end{align}

###Caso 2: Rama derecha ($U > 0.5$)

Para el intervalo $c < x \leq b$, la función de distribución acumulada se define como:
\begin{equation}
    F(x) = 1 - \frac{(b - x)^2}{(b - a)(b - c)}
\end{equation}

Igualamos a $U$ y sustituimos los valores correspondientes:
\begin{equation}
    U = 1 - \frac{(230 - x)^2}{(230 - 190)(230 - 210)} = 1 - \frac{(230 - x)^2}{(40)(20)} = 1 - \frac{(230 - x)^2}{800}
\end{equation}

Despejamos $x$ paso a paso:
\begin{align}
    \frac{(230 - x)^2}{800} &= 1 - U \nonumber \\
    (230 - x)^2 &= 800 \cdot (1 - U) \nonumber \\
    230 - x &= \sqrt{800 \cdot (1 - U)} \nonumber \\
    x &= 230 - \sqrt{800 \cdot (1 - U)}
\end{align}

### Ecuación Generadora Resultante

Por lo tanto, concluimos que la variable aleatoria $X$ (el peso de una tina individual) puede ser simulada a partir de un número aleatorio uniforme $U$ mediante la siguiente función definida a trozos:
\begin{equation}
    X =
    \begin{cases}
        190 + \sqrt{800 \cdot U} & \text{si } U \leq 0.5 \\
        230 - \sqrt{800 \cdot (1 - U)} & \text{si } U > 0.5
    \end{cases}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
n_simulaciones = 1000000  # Número de iteraciones (días simulados)
n_tinas = 5                 # Tinas por viaje
capacidad_camion = 1000     # Capacidad máxima en kg
```

```python
# Llevaremos un contador de cuántas veces el camión se pasa de la tonelada
veces_que_se_excede = 0
```

```python
# 2. Empezamos la simulación, repitiendo el proceso por cada día
for dia in range(n_simulaciones):
    peso_total_del_dia = 0
    # Subimos 5 tinas al camión
    for tina in range(n_tinas):
        U = random.random()
        if U <= 0.5:
            peso_tina = 190 + math.sqrt(800 * U)
        else:
            peso_tina = 230 - math.sqrt(800 * (1 - U))

        peso_total_del_dia = peso_total_del_dia + peso_tina
```

```python
if peso_total_del_dia > capacidad_camion:
    veces_que_se_excede = veces_que_se_excede + 1
```

```python
probabilidad = veces_que_se_excede / n_simulaciones
print("--- Resultados de la Simulación por Transformada Inversa ---")
print("Total de viajes simulados:", n_simulaciones)
print("Viajes que superaron los 1000 kg:", veces_que_se_excede)
print("Probabilidad de exceder la capacidad:", probabilidad * 100, "%")
```
