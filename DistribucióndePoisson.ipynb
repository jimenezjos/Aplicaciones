# Distribución de Poisson

## Demostración de la Identidad Recursiva de Poisson

Objetivo:

Demostrar que $p_{i+1} = \frac{\lambda}{i+1} p_i$ para $i \ge 0$.
1.  Partimos de la definición de la probabilidad para el valor $i+1$:
Sabemos que la función de masa de probabilidad de Poisson para cualquier valor $x$ es:
$$P\{X=x\} = e^{-\lambda} \frac{\lambda^x}{x!}$$
Si evaluamos esta función para $x = i+1$, obtenemos:$$p_{i+1} = e^{-\lambda} \frac{\lambda^{i+1}}{(i+1)!}$$

2. Descomponemos los términos exponenciales y factoriales:
Por las leyes de los exponentes, podemos separar $\lambda^{i+1}$ como:$$\lambda^{i+1} = \lambda \cdot \lambda^i$$Por la definición del factorial, podemos expandir $(i+1)!$ como:
$$(i+1)! = (i+1) \cdot i!$$

3. Sustituimos estas descomposiciones en la ecuación original:

$$p_{i+1} = e^{-\lambda} \frac{\lambda \cdot \lambda^i}{(i+1) \cdot i!}$$

4. Reagrupamos los términos:Ahora, separamos la fracción para aislar los términos que componen la probabilidad original $p_i$:
$$p_{i+1} = \frac{\lambda}{i+1} \left( e^{-\lambda} \frac{\lambda^i}{i!} \right)$$
5. Sustituimos la definición original de $p_i$:Observamos que el término dentro del paréntesis es exactamente la definición de $p_i$:$$p_i = e^{-\lambda} \frac{\lambda^i}{i!}$$Por lo tanto, al sustituir, llegamos a la identidad final:$$p_{i+1} = \frac{\lambda}{i+1} p_i$$

## Interpretación y Utilidad:

**Procesos Estocásticos:** Es el pilar matemático para modelar procesos de conteo y líneas de espera (teoría de colas), donde los eventos ocurren de manera continua e independiente.

**Modelado Financiero y Riesgo:** Se utiliza frecuentemente para modelar la ocurrencia de eventos discretos en carteras de inversión, como estimar la probabilidad de incumplimiento de pagos o modelar "saltos" repentinos en la volatilidad de los tipos de cambio.

**Análisis Deportivo:** Es sumamente útil para modelar y simular variables como el número de goles que un equipo anotará en un partido (por ejemplo, en la Liga MX), permitiendo calcular probabilidades de victoria basándose en promedios históricos.

```python
import random as rm
import math as mh
import matplotlib.pyplot as plt
import numpy
```

```python
def distribucion_poisson(lam):
    U = rm.random()  # 1. Generar U ~ U(0,1)
    i = 0
    p = mh.exp(-lam)
    F = p                # 2. Inicialización

    while True:
        if U < F:        # 3. Condición de término
            return i
        else:
            p = (lam * p) / (i + 1)  # 4. Actualización de probabilidad p
            F = F + p                # Actualización de probabilidad acumulada F
            i = i + 1                # Incremento de i
        # 5. El ciclo while actúa de forma natural como el "Ir al Paso 3"

# Ejemplo de uso
lam_ejemplo = 2.5
resultado = distribucion_poisson(lam_ejemplo)
print(f"Variable aleatoria generada (lambda = {lam_ejemplo}): X = {resultado}")
```
